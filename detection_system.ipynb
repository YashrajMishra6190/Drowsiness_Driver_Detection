```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import pygame
import time

# Load the pre-trained model
#model = load_model('drowsiness_detection_model.h5')
model = load_model('drowsiness_detection_model_augmented.keras')
# Initialize Pygame for playing sound
pygame.mixer.init()
pygame.mixer.music.load('emergency.mp3')

# Function to predict the state of the eyes (open or closed)
def predict_eye_state(frame):
    img = cv2.resize(frame, (64, 64))
    img = np.expand_dims(img, axis=0)
    img = img / 255.0
    prediction = model.predict(img)
    return prediction[0][0]  # Returns probability (close to 0 = closed, close to 1 = open)

# Capture video from the webcam
cap = cv2.VideoCapture(0)

# Run a loop to continuously capture frames from the webcam
while True:
    ret, frame = cap.read()
    
    if not ret:
        print("Failed to grab frame.")
        break
    
    # Convert the frame to grayscale (for face/eye detection)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Load a pre-trained frontal face detector from OpenCV
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        # Draw a rectangle around the face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # Region of interest for detecting eyes
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        # Detect eyes within the face
        eyes = eye_cascade.detectMultiScale(roi_gray)

        for (ex, ey, ew, eh) in eyes:
            # Draw a rectangle around the eyes
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

            # Crop the eye region from the frame for prediction
            eye = roi_color[ey:ey + eh, ex:ex + ew]

            # Predict if eyes are open or closed
            prediction = predict_eye_state(eye)

            # If prediction is less than 0.5, it indicates closed eyes
            if prediction < 0.5:
                cv2.putText(frame, "Drowsy - Eyes Closed!", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                pygame.mixer.music.play()
                time.sleep(3)  # Ensure alarm doesn't repeatedly trigger within a short time
            else:
                cv2.putText(frame, "Eyes Open", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                pygame.mixer.music.stop()

    # Display the resulting frame with detections
    cv2.imshow('Drowsiness Detection', frame)

    # Exit the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
